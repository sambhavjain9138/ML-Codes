```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_style('whitegrid')
sns.FacetGrid(data,hue="status",size=4).map(plt.scatter,"age","nodes").add_legend()
plt.show()
```

```python
sns.set_style('whitegrid')
sns.pairplot(data,hue="status",size=3)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt


sns.FacetGrid(data,hue="status",height=5).map(sns.histplot,"nodes").add_legend()
plt.show
```

```python
sns.boxplot(x='status',y='nodes',data=data)
plt.show()
```

```python
#2D Density plot, contors-plot
sns.jointplot(x="age", y="nodes", data=data, kind="kde");
plt.show();
```
