```python
import matplotlib.pyplot as plt
plt.plot([2,4,6,4])
plt.xlabel("index")
plt.ylabel("value")
plt.grid()        #to show grid on the graph
plt.show()      #to produce final graph
```

```python
import matplotlib.pyplot as plt
plt.plot([1,2,4,6,4],[0,4,16,36,16])
plt.xlabel("index")
plt.ylabel("value")
plt.grid()        #to show grid on the graph
plt.show()      #to produce final graph
```

```python
#to plot multiple graph, we use different symbols and color
import matplotlib.pyplot as plt
plt.plot([2,4,6],[4,16,36],'b^')
plt.xlabel("index")
plt.ylabel("value")
plt.grid()        #to show grid on the graph
plt.show()      #to produce final graph
```

```python
import numpy as np
t=np.arange(0,5,0.5)
plt.plot(t,t**2,'b--',label='^2');
plt.plot(t,t**3,'bs',label='^3');
plt.plot(t,t**4,'b^',label='^4');
plt.grid()
plt.legend()
plt.show()
```

```python
plt.subplot(321)
plt.plot([2,4,6],[4,16,36],'b^')
plt.subplot(313)
plt.plot([2,4,6],[4,16,36],'b--')
plt.show()
```
